```python
# Ans => Matplotlib is a widely-used Python library for creating static, animated, and interactive visualizations in a variety of formats. It provides a comprehensive 
# and flexible set of tools for creating various types of plots and charts, making it a powerful tool for data visualization and analysis.

# *Matplotlib is used for:-

# 1.Data Visualization:- Matplotlib allows you to create a wide range of static and interactive visualizations, making it essential for data scientists, researchers, 
# and analysts to explore and communicate data effectively.

# 2.Publication-Quality Graphics:- It provides fine-grained control over the appearance of plots, enabling users to create publication-quality graphics for research
# papers, reports, and presentations.

# 3.Customization:- Matplotlib's flexibility allows users to customize nearly every aspect of a plot, including colors, labels, markers, fonts, and more.

# 4.Integration with Other Libraries:- It integrates seamlessly with other Python libraries like NumPy, Pandas, and SciPy, making it an essential tool in the data 
# science ecosystem.

# 5.Support for Multiple Plot Types:- Matplotlib supports a wide range of plot types. Here are five common plots that can be created using the Pyplot module of 
# Matplotlib:

# a. Line Plot: Line plots are used to visualize the relationship between two continuous variables, typically over a continuous range.

# b. Scatter Plot: Scatter plots are used to display individual data points as markers on a 2D plane. They are helpful for visualizing the distribution and 
# relationships between two variables.

# c. Bar Plot: Bar plots are used to represent categorical data with rectangular bars. They are often used to compare values across categories.

# d. Histogram: Histograms are used to visualize the distribution of a single variable. They divide the data into bins and display the frequency or density of each bin.

# e. Pie Chart: Pie charts are circular charts used to represent the composition of a whole by dividing it into sectors. Each sector represents a category's proportion
# in the whole...
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# import numpy as np
# company = np.array(["Apple", "Microsoft", "Google", "AMD"])
# profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Ans => A bar plot, also known as a bar chart or bar graph, is a data visualization technique that uses rectangular bars or 
# columns to represent categorical data. Each bar's length or height is proportional to the value it represents. Bar plots
# are commonly used to display and compare data between different categories or groups. They are used for the following purposes:-
# 1.Easy to Understand:- Bar plots are easy to understand and interpret, making them suitable for both technical and non-technical 
# audiences.
# 2.Horizontal Bar Plots:- Horizontal bar plots are a variation of bar plots where the bars are oriented horizontally instead of
# vertically. They are often used when you have long category names or when you want to emphasize the ranking or comparison of 
# categories.
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(8,4))
plt.bar(company,profit)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(company,profit)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# box1 = np.random.normal(100, 10, 200)
# box2 = np.random.normal(90, 20, 200)
```

```python
# Ans => A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset.
# It is used to visualize the spread and central tendency of the data, as well as identify potential outliers.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1,box2])
plt.show()
```
